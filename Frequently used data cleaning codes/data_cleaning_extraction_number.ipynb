- In columns where units exist together with numbers,  
  such as 'Volume', 'Weight', and 'Length',  
  there are cases where you want to separate only the numbers for a specific calculation.

- In this case, I use a regular expression to extract the number.

```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_excel("data/sample_data_volume.xlsx")
print(df.shape)
df
```

```python
df['volume_1'] = np.where(df['volume'].notnull(), df['volume'].apply(lambda x: re.sub(r'[^\d*.?\d+]', '', str(x))), np.nan)
```
